```python
import fnmatch
import os
import arcpy
import pandas as pd

```

```python
def get_path_for_all_feature_from_gdb(input_gdb, type ="Polygon", data_type="FeatureClass"):
    gdbPath =[]
    for dirpath, dirnames, filenames in arcpy.da.Walk(input_gdb, datatype=data_type, type=type):
        for filename in filenames:
            gdbPath.append(os.path.join(dirpath, filename))
    return gdbPath

def return_merged_tables_from_fc_list(in_gdb, field):

    fc_list = get_path_for_all_feature_from_gdb(in_gdb, type='POLYLINE')
    df_list=[]
    for x in fc_list:

        df_list.append(pd.DataFrame(arcpy.da.FeatureClassToNumPyArray(x, field)))

    df_master = pd.concat(df_list)

    return df_master
```

```python
df = return_merged_tables_from_fc_list(r"D:\Census_Data\US Map\Roads\Road_by_state_by_grid_diss.gdb", field= ['id','STATE_FIPS', 'MTFCC', 'road_miles'])
```

```python
df_road_mile_pivot = pd.DataFrame(df.loc[df.drive_test_road==1].pivot(index ="id", columns='MTFCC', values='road_miles'))
```

```python
df_road_mile_pivot.reset_index(inplace=True)
```

```python
df4 = pd.merge(left = df2, left_on='grid_id', right = df_road_mile_pivot, right_on='id', how='outer')
```
